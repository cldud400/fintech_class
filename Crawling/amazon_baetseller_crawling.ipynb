```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
path = '#zg-right-col'
contents = soup.select(path)

for i in range(10):
    # 랭킹
    rank = contents[0].select('div.a-section.zg-bdg-body.zg-bdg-clr-body.aok-float-left > span')[i].text
    # 타이틀
    title = contents[0].select('span > div')[i].text
    # 댓글수
    num_review = contents[0].select('div.zg-grid-general-faceout > div > div > div > a > span')[i].text
    # 가격
    try:
        price = contents[0].select('div:nth-child({}) > a > span > span > span'.format(i+1))[0].text
    except:
        price = contents[0].select('div:nth-child({}) > a > span'.format(i+1))[0].text


    print('랭킹: ', rank)
    print('타이틀: ', title)
    print('댓글수: ', num_review)
    print('가격: ', price)
    print('=' * 100)
```
