```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.imdb.com/movies-coming-soon/2022-02/'
```

```python
# 내가 한 것
res = requests.get(url)
html = res.text
soup = BeautifulSoup(html, 'html.parser')
path = '#main > div > div.list.detail > div'
lis = soup.select(path)
for li in lis:
    genre = []
    names = []
    title_ = li.select('td > h4 > a')[0].text.strip()
    for i in li.select('p > span'):
        if i.text == '|':
            continue
        elif i.text.isdecimal():
            continue
        genre.append(i.text)
    content = li.select('div.outline')[0].text.strip()
    director = li.select('span > a')[0].text
    try:
        time = li.select('time')[0].text
    except:
        pass
    for i in li.select('div.txt-block > a'):
        names.append(i.text)
    print('Title:', title_)
    print('Genre:', genre)
    print('Contect:', content)
    print('Director:', director)
    print('Stars:', names)
    print('Time:',time)
    print('-' * 100)
```

### dict에 담아서 Pandas로 출력   

---

```python
res = requests.get(url)
html = res.text
soup = BeautifulSoup(html, 'html.parser')
path = '#main > div > div.list.detail > div'
lis = soup.select(path)
movies = []
for li in lis:
    genre = []
    names = []
    title_ = li.select('td > h4 > a')[0].text.strip()
    for i in li.select('p > span'):
        if i.text == '|':
            continue
        elif i.text.isdecimal():
            continue
        genre.append(i.text)
    content = li.select('div.outline')[0].text.strip()
    director = li.select('span > a')[0].text
    try:
        time = li.select('time')[0].text
    except:
        pass
    for i in li.select('div.txt-block > a'):
        names.append(i.text)
    d = dict()
    d['Title'] = title_
    d['Genre'] = genre
    d['Content'] = content
    d['Director'] = director
    d['Stars'] = names
    d['Time'] = time
    movies.append(d)
```

```python
import pandas as pd
movies_df = pd.DataFrame(movies)
movies_df.to_csv('movies.csv')
```
