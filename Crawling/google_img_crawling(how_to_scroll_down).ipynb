## 페이지 스크롤을 내리는 방법 ( javascript )
---
- driver.excute_script() : 자바 스크립트 코드를 사용할 수 있게 해 준다.

```python
last_height = driver.execute_script("return document.body.scrollHeight")
SCROLL_PAUSE_TIME = 1


while True:
    driver.execute_script('window.scrollTo(0, document.body.scrollheight)')
    time.sleep(SCROLL_PAUSE_TIME)
    
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        try:
            driver.find_element_by_css_selector(".mye4qd").click()
        except:
            break
            
    last_height = new_height
```

```python
# 강사님
url = 'https://www.google.co.kr/'
driver = webdriver.Chrome()
time.sleep(0.5)
driver.get(url)
time.sleep(0.5)
elem = driver.find_element_by_css_selector('input.gLFyf.gsfi')
time.sleep(0.5)
elem.send_keys('빅데이터')
time.sleep(0.5)
elem.send_keys(Keys.RETURN)
time.sleep(0.5)
driver.find_element_by_xpath('//*[@id="hdtb-msb"]/div[1]/div/div[2]/a').click()
time.sleep(0.5)
images = driver.find_elements_by_css_selector('img.rg_i.Q4LuWd')


for i, img in enumerate(images):
    img.click()
    elem = driver.find_element_by_xpath('//*[@id="islrg"]/div[1]/div[{}]/a[1]/div[1]/img'.format(i+1))
    img_url = elem.get_attribute('src')
#     img_url = img.get_attribute('src')
    urllib.request.urlretrieve(img_url, './google_img_folder/google{}.jpg'.format(i))
    if i == 10:
        break
```
