```python
# groupby
# 동일한 기준을 갖는 데이터끼리 그룹화 시켜준다
# 1. 데이터를 묶는 과정 => 그룹생성
# 2. 그룹별 어떤 값을 출력? => 그룹별 연산과정
```

```python
df = pd.read_csv('./data/gapminder.tsv', sep = '\t')
```

```python
df.head()
```

```python
df.groupby('year')['lifeExp'].mean()
```

```python
grouped = df.groupby('year')
grouped.lifeExp.mean()
```

```python
# grouped 객체에 적용되는 메소드
# agg - 함수를 적용하는 메소드, 사용자 정의함수를 적용시킬 때 사용
# filter - 원하는 그룹만 추출
# transform - 데이터의 값을 그룹의 값으로 변경
```

```python
# agg - 사용자 정의함수, 다양한 함수를 적용, 칼럼별 함수 적용
```

```python
grouped = df.groupby('year')
grouped.lifeExp.agg(np.mean)
```

```python
def mean_dff(values, diff_value):
    n = len(values)
    sum_ = 0
    for value in values:
        sum_ += value
    mean = sum_/n
    return mean - diff_value     # 평균과의 차이를 계산
```

```python
global_mean = df.lifeExp.mean()  # 데이터의 평균
```

```python
df.groupby('year').lifeExp.agg(mean_dff, diff_value=global_mean)    # 편차 계산
```

```python
df.groupby('year').lifeExp.agg([np.mean, np.std, np.count_nonzero])
```

```python
df.groupby('year').agg({'lifeExp':'mean', 'pop':'median', 'gdpPercap':'max'})
```

```python
# Transforn - 그룹별 값으로 데이터 값을 바꾼다
```

```python
def my_zscore(x):
    return (x-x.mean())/x.std()
```

```python
df.groupby('year').lifeExp.transform(my_zscore)
```

```python
df['z_score'] = df.groupby('year').lifeExp.transform(my_zscore)
df.head(3)
```

```python
tips10 = sns.load_dataset('tips').sample(10)
tips10.loc[np.random.permutation(tips10.index)[:4], 'total_bill'] = np.NaN
tips10
```

```python
def fill_na_mean(x):
    avg = x.mean()
    return x.fillna(avg)
```

```python
tips10.groupby('sex').total_bill.transform(fill_na_mean)
```
