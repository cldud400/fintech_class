## grouped

```python
# 월별 통신 이용시간의 합을 계산
# grouped = phone.groupby(['month'])
# grouped.agg(sum)

phone.groupby(['month'])['duration'].sum()
```

```python
# 월별 통화 시간의 합
phone[phone.item=='call'].groupby('month')['duration'].sum()
```

```python
# 아이템별 이용시간 합
phone.groupby('item')['duration'].sum()
```

```python
# 월별, 아이템별 이용건수
phone.groupby(['month','item'])['date'].count()
```

```python
phone.groupby(['month','item'])['date'].count().unstack()  # index가 중첩이 될 때 unstack을 하여 테이블로 만들 수 있다.
```

## merge

```python
person = pd.read_csv('./pandas_datafiles/survey_person.csv')
site = pd.read_csv('./pandas_datafiles/survey_site.csv')
survey = pd.read_csv('./pandas_datafiles/survey_survey.csv')
visited = pd.read_csv('./pandas_datafiles/survey_visited.csv')
```

```python
site.head(1)
```

```python
visited.head(1)
```

```python
# site, visited를 inner merge
# left_on, right_on 다른 col인데 실질적으로 내용이 같을 때
pd.merge(site,visited, left_on = 'name', right_on='site', how = 'inner')
```

```python
person.head(1)
```

```python
survey.head(1)
```

```python
# person, survey를 inner merge
pd.merge(person, survey, left_on = 'ident', right_on = 'person')   # how : default 값이 'inner'라 생략가능
```
