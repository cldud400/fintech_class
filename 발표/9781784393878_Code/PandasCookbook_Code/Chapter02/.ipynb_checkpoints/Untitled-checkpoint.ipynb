# 연산

```python
college = pd.read_csv('../data/college.csv', index_col = 'INSTNM')
college.head()
```

```python
# college + 5

# object 타입의 데이터가 포함이 되어 있기 때문에 연산이 안됨

# 데이터 타입을 선택하자
# filter로 선택
collf = college.filter(like='UGDS_')
collf.head()

# type select로 선택
collts = college.select_dtypes('number')
colls + 5
```

반올림

```python
collf
```

```python
# 소수점 3번째 자리에서 반올림
collf.round(2)
# 단, 예)0.0250은 0.02가 되기 때문에 + 0.0001을 하는 것이 좋다
(collf+0.0001).round(2)
```

```python
# 연산자 메서드
# floordiv : x // y (몫)
collf.add(0.0001).floordiv(0.01).div(100)
```

연산 방향 전환
---
axis() 사용

```python
coll = pd.read_csv('../data/college.csv', index_col='INSTNM')
coll = coll.filter(like='UGDS_')
coll.head()
```

```python
coll.count(axis = 0)
coll.count(axis = 1)
```

```python
coll.sum(axis = 0) == coll.sum(axis='index')
```

```python
# cumsum : 누적 백분율
coll_cumsum = coll.cumsum(axis=1)
coll_cumsum
```

```python
# 내림차순 정렬
coll.iloc[0].sort_values(ascending=False)
```

drop()

```python
# 모든 Nan값을 drop하면 다양한 계산이 가능
# any: 모든 행의 컬럼값 중 하나라도 NULL의 값을 가지면 해당 행을 제거
# all: 모든 칼럼 값이 NULL이거나 NaN인 경우에만 해당 행을 제거
coll = coll.dropna(how='all')

coll.isnull().sum()
```

ge()

```python
# .ge() : () 보다 크면 True, 작으면 False
coll.ge(0.15)

# ()보다 큰 수의 갯수를 카운트 할 수있다.
coll.ge(0.15).sum()

# index에 ()보다 큰 값이 몇개가 있는가??
coll.ge(0.15).sum(axis = 'columns')
coll.ge(0.15).sum(axis = 'columns').sort_values(ascending=False).head()
```
