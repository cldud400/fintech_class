```python
# isinstance(x, type명) : x의 type이 type명과 일치하는지 확인한다
# data_dict의 value가 str이면 '', 아니면 nan을 리턴
new_data_dict = {k: '' if isinstance(v, str) else
    np.nan for k, v in data_dict.items()}
new_data_dict
```

### How it works...

```python
# value가 str이면 abcde중 아무거나
# value가 str이 아니면 1~10중 아무거나
random_data = []
for i in range(1000):   # doctest: +SKIP
    d = dict()
    for k, v in data_dict.items():
        if isinstance(v, str):
            d[k] = np.random.choice(list('abcde'))
        else:
            d[k] = np.random.randint(10)
    random_data.append(pd.Series(d, name=i + len(bball_16)))
random_data[0]
```

```python
%%timeit
bball_16_copy = bball_16.copy()
for row in random_data:
    bball_16_copy = bball_16_copy.append(row)
```

```python
%%timeit
bball_16_copy = bball_16.copy()
bball_16_copy = bball_16_copy.append(random_data)
```

## concat, join, merge의 차이점
---
### concat
- pandas의 함수
- 둘 이상의 pandas객채를 수직이나 수평으로 병합
- 인덱스에 대해서만 정렬
- 기본설정은 outer join, inner join은 옵션으로 지정해 줘야한다.

### .jion
- DataFrame 메서드
- 둘 이상의 pandas 객체를 수평으로 병합
- 호출 DataFrame의 열 또는 인덱스를 다른 객체의 인덱스에 대해 정렬
- 합쳐지는 열/인덱스에 중복값이 있으면 카티션 곱을 생성
- 기본설정은 left join이고, 옵션 설정을 통해 inner join, outer join, right join가능

### .merge
- DataFrame 메서드
- 정확히 두 DataFrame을 수평으로 조인
- 호출 DataFrame의 열 또는 인덱스를 다른 DataFrame의 열 또는 인덱스와 정렬
- 합쳐지는 열/인덱스에 중복값이 있으면 카티션 곱을 생성
- 기본설정은 inner join이고, 옵션 설정을 통해 outer join, left join, right join가능

```python
from IPython.display import display_html
years = 2016, 2017, 2018
stock_tables = [pd.read_csv(
    r'C:\Users\bitcamp\Desktop\data/stocks_{}.csv'.format(year), index_col='Symbol')
    for year in years]
stocks_2016, stocks_2017, stocks_2018 = stock_tables
stocks_2016
```

```python
pd.concat(stock_tables, keys=[2016, 2017, 2018])
```

```python
# dict(zip(years, stock_tables)) : year을 key, stock_tables를 value
pd.concat(dict(zip(years, stock_tables)), axis='columns')
```

```python
# 동일한 이름의 열이 존재하면 lsuffix와 rsuffix로 이름 지정 가능
stocks_2016.join(stocks_2017, lsuffix='_2016',
    rsuffix='_2017', how='outer')
```

```python
# add_suffix('_2017')로 열의 이름을 변경하여 outer join
other = [stocks_2017.add_suffix('_2017'),
    stocks_2018.add_suffix('_2018')]
stocks_2016.add_suffix('_2016').join(other, how='outer')
```

```python
# 
stock_join = stocks_2016.add_suffix('_2016').join(other,
    how='outer')
stock_concat = pd.concat(dict(zip(years,stock_tables)),
    axis='columns')
level_1 = stock_concat.columns.get_level_values(1)
level_0 = stock_concat.columns.get_level_values(0).astype(str)
# concat의 열 이름에 접미어 넣기
stock_concat.columns = level_1 + '_' + level_0
stock_join.equals(stock_concat)
```

```python
stock_concat = pd.concat(dict(zip(years,stock_tables)),
    axis='columns')
stock_concat
```

```python
# merge는 자동으로 접미어를 제공
stocks_2016.merge(stocks_2017, left_index=True,
    right_index=True)
```

```python
# merge는 자동으로 인덱스 정렬을 해주기 때문에 concat와 equals를 해보면 False
step1 = stocks_2016.merge(stocks_2017, left_index=True,
    right_index=True, how='outer',
    suffixes=('_2016', '_2017'))
stock_merge = step1.merge(stocks_2018.add_suffix('_2018'),
    left_index=True, right_index=True,
    how='outer')
stock_concat.equals(stock_merge)
```

### How it works...

```python
sql_string1 = '''
SELECT
    Name,
    time(avg(Milliseconds) / 1000, 'unixepoch') as avg_time
FROM (
      SELECT
          g.Name,
          t.Milliseconds
      FROM
          genres as g
      JOIN
          tracks as t on
          g.genreid == t.genreid
     )
GROUP BY Name
ORDER BY avg_time'''
pd.read_sql_query(sql_string1, engine)
```

```python
sql_string2 = '''
   SELECT
         c.customerid,
         c.FirstName,
         c.LastName,
         sum(ii.quantity * ii.unitprice) as Total
   FROM
        customers as c
   JOIN
        invoices as i
        on c.customerid = i.customerid
   JOIN
       invoice_items as ii
       on i.invoiceid = ii.invoiceid
   GROUP BY
       c.customerid, c.FirstName, c.LastName
   ORDER BY
       Total desc'''
```

```python
pd.read_sql_query(sql_string2, engine)
```
