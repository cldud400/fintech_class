```python
res = requests.get('http://corners.gmarket.co.kr/Bestsellers')
soup = BeautifulSoup(res.content, 'html.parser')

data_dict = dict()

categories = soup.select('div.gbest-cate ul.by-group li a')

category_name_list = []
sub_category_name_list = []
ranking_list = []
title_list = []
old_price_list = []
dis_price_list = []
discount_percent_list = []
item_code_list = []
provider_list = []

base_url = 'http://corners.gmarket.co.kr'



for category in categories:
    cate_url = 'http://corners.gmarket.co.kr/' + category['href']
    cate_title = category.text
    
    print('='*30)
    print()
    print(cate_title)
    
    res1 = requests.get(cate_url)
    soup1 = BeautifulSoup(res1.content, 'html.parser')
    
    atags = soup1.select('div.navi.group > ul > li > a')
    
    for a in atags:
#         print(a.text)
#         print(base_url + a['href'])
        sub_url = base_url + a['href']
        sub_title = a.text
        
        print()
        print(sub_title)
        
        res = requests.get(sub_url)
        soup = BeautifulSoup(res.content, 'html.parser')

        prods = soup.select('div.best-list > ul > li')

        for i in range(5,11):     # 5 ~105
#             ranking = prods[i].select('li > p')[0].text
#             title = prods[i].select('li > a.itemname')[0].text
#             prod_url = prods[i].select('li > a.itemname')[0]['href']
#             old_price = prods[i].select('li > div.item_price > div.o-price')[0].text.replace(',','')[:-1]
#             dis_price = prods[i].select('li > div.item_price > div.s-price')[0].text.split('\n')[1].replace(',','')[:-1]
#             dis_per = prods[i].select('li > div.item_price > div.s-price')[0].text.split('\n')[2][:-1]

#             res = requests.get(prod_url)
#             soup = BeautifulSoup(res.content, 'html.parser')

#             provider = soup.select('div.item-topinfo > div > p > a > strong')[0].text
#             prod_code = soup.select('div.item-topinfo > div > p > span.rt > span.pdnum')[0].text.split(':')[1].strip()
#             prod_code
            
            try:
                ranking = prods[i].select('li > p')[0].text
                title = prods[i].select('li > a.itemname')[0].text
                prod_url = prods[i].select('li > a.itemname')[0]['href']
                old_price = prods[i].select('li > div.item_price > div.o-price')[0].text.replace(',','')[:-1]
                dis_price = prods[i].select('li > div.item_price > div.s-price')[0].text.split('\n')[1].replace(',','')[:-1]
                dis_per = prods[i].select('li > div.item_price > div.s-price')[0].text.split('\n')[2][:-1]

                res = requests.get(prod_url)
                soup = BeautifulSoup(res.content, 'html.parser')

                try:
                    provider = soup.select(' div.item-topinfo_headline > p > span.text__brand > span.text')[0].text
                except:
                    provider = []
                try:
                    prod_code = soup.select('body > div.location-navi > div > span')[0].text.split(':')[1].strip()
                except:
                    prod_code = []

                print(ranking,title,old_price,dis_price,dis_per,provider, prod_code)
            except:
                pass

            category_name_list.append(cate_title)
            sub_category_name_list.append(sub_title)
            ranking_list.append(ranking)
            title_list.append(title)
            old_price_list.append(old_price)
            dis_price_list.append(dis_price)
            discount_percent_list.append(dis_per)
            item_code_list.append(prod_code)
            provider_list.append(provider)
            
#             print(ranking,title,old_price,dis_price,dis_per,provider, prod_code)
            
# data_dict['category_name'] = category_name_list
# data_dict['sub_category_name'] = sub_category_name_list
# data_dict['ranking'] = ranking_list
# data_dict['title'] = title_list
# data_dict['ori_price'] = old_price_list
# data_dict['dis_price'] = dis_price_list
# data_dict['discount_percent'] = discount_percent_list
# data_dict['item_code'] = item_code_list
# data_dict['provider'] = provider_list
```

```python
sql = '''
use gmarketdb;
'''
cursor.execute(sql)
```

```python
sql = '''
create table ranking(
    main_category varchar(20),
    sub_category varchar(20),
    item_ranking int,
    item_code varchar(20),
    privider varchar(20));
'''
cursor.execute(sql)
```

```python
sql = '''
create table item(
    item_code varchar(20) not null primary key,
    title varchar(20),
    old_price int,
    dis_price varchar(20),
    dis_percent int);
'''
cursor.execute(sql)
```
