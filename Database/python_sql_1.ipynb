### groupby 객체 메소드 : agg, filter, transform
---
1. agg : groupby 객체에 함수를 적용하는 메소드 (apply와 비슷)
2. filter : 특정 group을 추출
3. transform : 그룹의 값으로 데이터값을 바꿈

```python
# agg : column별로 다른 함수 적용, 여러개의 함수 적용이 가능하다
payment_df.groupby('month').agg({'payment_id' : np.size, 'amount' : [np.sum, np.mean]})
```

```python
# sql 버전
sql = '''
select count(payment_id) '구매건수', avg(amount) '평균구매액', sum(amount) '총구매액', left(payment_date, 7) as 'month'
    from payment group by 'month'
'''
payment_df = pd.read_sql(sql,db)
payment_df
```
