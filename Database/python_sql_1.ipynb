```python
sql = '''
show tables;
'''
cursor.execute(sql)

results = cursor.fetchall()
for file in results[1]:
    name_ = file[0]
    sql = 'select * from {}'.format(file[0])
    table_df = pd.read_sql(sql,db)
    
    table_df.to_csv('./sakila/{}.csv'.format(name_))
    
```

```python
# 강사님
sql = '''
show tables;
'''
tables = pd.read_sql(sql, db)
```

```python
for t in tables.values:
    print(t[0])
    
    sql = 'select * from {}'.format(t[0])
    table_df = pd.read_sql(sql,db)
    table_df.to_csv('./sakila/{}.csv'.format(t[0]))
```
