## NaN값 처리하기

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(5,5), index=['a','b','c','d','e'],
             columns=['one','two','three','four','five'])             # 표준정규분포에서 5^5로 변수를 뽑는다
```

```python
df
```

```python
df.iloc[1:3, 1] = np.nan
df.iloc[2, 2:4] = np.nan
df
```

## Nan, NaN, nan 처리방식
---
0. nan값을 찾는다  => isnull(), isna()
1. nan값을 포함한 row 또는 column을 삭제  => drop()
2. nan값을 다른 값으로 채우는 방법  => fillna()

> 0. nan값을 찾는다

```python
df.isnull()    # nan이면 True 아니면 False
```

```python
df.isnull().sum()    # df에 함수를 적용하면 기분 axis = 0(세로방향)
```

```python
df.isnull().sum().sum()
```

```python
df.isnull().sum(axis = 1)
```

> 1. nan값을 포함한 row 또는 column을 삭제  => drop()

```python
df.dropna()     # 기본으로 row를 삭제 dorp : (axis = 0 : 가로 삭제)
```

```python
df.dropna(axis = 1)
```

```python
df
```

```python
df.dropna(how = 'all')    # 전체 데이터가 nan인 경우 삭제
```

```python
df.dropna(thresh = 3)    # 데이터가 3개 이상이면 삭제하지 마라
```

> 2. nan값을 다른 값으로 채우는 방법  => fillna()

```python
df
```

```python
df.fillna(method = 'ffill')   # 뒤의 데이터로 채우기 (세로 기준 위)
```

```python
df.fillna(method = 'bfill')      # 뒤의 데이터로 채우기 (세로 기준 밑)
```

```python
df.fillna(method = 'bfill', axis = 1)     # 뒤의 데이터로 채우기 (가로 기준 오른쪽)
```

```python
df.fillna(df.mean())    # nan값을 평균값으로 채워라 (세로방향의 평균)
```

```python
df.interpolate()       # 보간법(내삽) : nan값을 데이터 근삿값으로 채워라
```

# groupby
---
- 특정 기준값으로 동직적인 데이터를 묶는 방법

```python
data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'Giant', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(data)
df
```

```python
df.head()              # 상위 5개
```

```python
# df.groupby('Team')         # 팀을 기준으로 데이터를 묶는다 => grouping
# ['Points'].mean() - group의 값을 어떤 값으로 표현할지??

df.groupby('Team')['Points'].mean()   # 데이터를 묶은 후 Points의 평균값으로 표현
```

```python
# Team별로 가장 좋은 ranking를 뽑는다
df.groupby('Team')['Rank'].min()
```

```python
# 연도별 최고 점수
```

```python
df.groupby('Year')['Points'].max()
```

```python
# 팀별 데이터 수는 각각 몇개?
# df.groupby('Team').count()
df.groupby('Team')['Points'].count()
```

```python
# 팀별, 연도별 포인트 평균
df.groupby(['Team','Year'])['Points'].mean()
```

## grouped()

```python
grouped = df.groupby('Team')
```

```python
grouped.agg(min)      # 모든 데이터를 group화해서 최소값만 출력
```

```python
grouped.agg(np.mean)
```

```python
grouped.agg([np.sum, np.mean, np.std])
```

```python
grouped.get_group('Riders')        # group에서 'Riders'의 데이터만 가져옴
```

# merge / concat
---
- dataframe를 합치는 것
- merge : sql의 table을 합하는 join과 같다
![merge_img](https://data36.com/wp-content/uploads/2018/08/4-pandas-merge-inner-outer-left-right-768x579.png)

## merge의 종류   
---
A_df와 B_df를 병합하는 경우

1. inner merge : A_df와 B_df 모두에 존재하는 데이터(row)들로만 합하는 방식
2. outer merge : A_df와 B_df 어느 한 곳에만 존재하는 데이터를 모두 포함해서 합하는 방식
3. left merge : A_df의 데이터를 모두 포함, B_df에서 A_df에 존재하는 데이터만 포함
4. right merge : B_df의 데이터를 모두 포함, A_df에서 B_df에 존재하는 데이터만 포함

```python
raw_data = {
    'subject_id' : [str(i) for i in range(1,12) if i != 6],
    'test_score' : [51,15,15,61,16,14,15,15,16,61]
}
df_a = pd.DataFrame(raw_data)
df_a
```

```python
raw_data = {
    'subject_id' : ['4','5','6','7','8'],
    'first_name' : ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
    'last_name' : ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']
}
df_b = pd.DataFrame(raw_data)
df_b
```

```python
df_a.subject_id.values
```

```python
df_b.subject_id.values
```

```python
# pd.merge(df_a, df_b, on : 기준이 되는 칼럼, how : 4가지 방법 (default : 'inner'))
merged_df = pd.merge(df_a, df_b, on='subject_id', how = 'inner')
merged_df
```

```python
# outmerge : nan값 생성 가능성이 크다
pd.merge(df_a, df_b, on='subject_id', how = 'outer')
```

```python
df_a.merge(df_b, on = 'subject_id', how = 'left')
```

```python
df_a.merge(df_b, on = 'subject_id', how = 'right')
```

## concat
---
- 두 개 이상의 dataframe을 물리적으로 병합

```python
raw_data = {
    'subject_id' : [str(i) for i in range(1,12) if i != 6],
    'test_score' : [51,15,15,61,16,14,15,15,16,61]
}
df_a = pd.DataFrame(raw_data)
df_a
```

```python
raw_data = {
    'subject_id' : ['4','5','6','7','8'],
    'first_name' : ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
    'last_name' : ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']
}
df_b = pd.DataFrame(raw_data)
df_b
```

```python
pd.concat([df_a,df_b])
```

```python
pd.concat([df_a,df_b], axis = 1)
```
