```python
import pandas as pd
import numpy as np

phone = pd.read_csv('./pandas_datafiles/phone_data.csv', index_col = 0)     # 0번 col을 index로 해라
```

```python
phone.head(10)
```

```python
phone.info()
```

```python
phone.isnull().sum().sum()    # nan값이 없다
```

```python
phone.describe()        # column이 숫자일 경우에만 평균, 최대, 최소 등등 출력
```

```python
phone['date'] = pd.to_datetime(phone['date'])      # date row를 날짜 형식으로 바꿔준다
phone
```

```python
# 월별 통신 이용시간의 합을 계산
# grouped = phone.groupby(['month'])
# grouped.agg(sum)

phone.groupby(['month'])['duration'].sum()
```

```python
# 월별 통화 시간의 합
phone[phone.item=='call'].groupby('month')['duration'].sum()
```

```python
# 아이템별 이용시간 합
phone.groupby('item')['duration'].sum()
```

```python
# 월별, 아이템별 이용건수
phone.groupby(['month','item'])['date'].count()
```

```python
phone.groupby(['month','item'])['date'].count().unstack()  # index가 중첩이 될 때 unstack을 하여 테이블로 만들 수 있다.
```
