# Line chart
---
- 데이터의 추세를 확인
- plot()

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.plot(x,y)
```

```python
x = [1,2,3,4,5]
y = [i**2 for i in x]

x1 = [1,2,3,4,5]
y1 = [6,7,8,9,10]

plt.plot(x,y)
plt.plot(x1,y1)

# 그래프를 jpg파일로 저장하기
plt.savefig('text_vis.jpg')
```

```python
# 코사인함수
x1 = range(50)
y1 = [np.cos(value) for value in x1]

# 사인함수
x2 = range(50)
y2 = [np.sin(value) for value in x2]

plt.plot(x1,y1)
plt.plot(x2,y2)
```

```python
# 출력 사이즈 조절
plt.figure(figsize = (10,5))

# 출력을 1행 2열로 나눠서 첫번째에 출력
plt.subplot(1,2,1)
# 컬러와 라인스타일 변경 가능
plt.plot(x1,y1, color = 'red', linestyle = 'dotted')

# 출력을 1행 2열로 나눠서 두번째에 출력
plt.subplot(1,2,2)
# 컬러와 라인스타일 변경 가능
plt.plot(x2,y2, color = 'blue', linestyle = 'dashed')
```

```python
x1 = range(100)
y1 = [value for value in x1]

x2 = range(100)
y2 = [value+50 for value in x2]

# c: color, rgb코드 입력 가능      ls : linestyle         label : 그래프의 이름
plt.plot(x1,y1, c = '#a52a2a', ls = '-.', label = 'Line1')
plt.plot(x2,y2, c = '#0000ff', ls = ':', label = 'Line2')

# text 넣기  (50,60위치에, 'text')
plt.text(50,60, 'Line 1')

# 화살표(), text(30,110)위치에서 (50,100)을 가르킨다
plt.annotate('Line 2', xytext = (30,110), xy = (50,100), arrowprops = dict(facecolor='yellow', shrink = 0.1))

# label의 범례를 표시(위치값 세팅 , default : 빈공간)
plt.legend(fancybox = True, shadow = True, loc = 'upper right')

# 가로 세로 이름 붙이기
plt.xlabel('X_axis')
plt.ylabel('Y_axis')
```

# Scatter Plot
---
- 데이터의 분포
- scatter()

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.scatter(x,y)
```

```python
# 0~1까지의 수 중에 난수100개
x = np.random.rand(100)

# 표준정규분포 중 난수 100개 / 10
y = x + np.random.randn(100)/10

plt.scatter(x,y)
```

```python
x = np.random.rand(500)
y = x + np.random.randn(500)/10

plt.scatter(x,y)
```

```python
# 512 by 2
data1 = np.random.rand(512,2)
data2 = np.random.rand(512,2)
```

```python
# data1의 [:,0]을 x축, [:,1]을 y축, color = blue, marker = 'x'
plt.scatter(data1[:,0],data1[:,1],c='b', marker='x')
# data2의 [:,0]을 x축, [:,1]을 y축, color = red, marker = 'o'
plt.scatter(data2[:,0],data2[:,1],c='r', marker='o')
```

```python
n = 50
x = np.random.rand(n)
y = np.random.rand(n)

colors = np.random.rand(n)

# random 면적
area = np.pi * (30 * np.random.rand(n))**2

# s = size,    alpha = 투명도
plt.scatter(x,y, c=colors, s=area, alpha = 0.5)
# 이미지 저장
plt.savefig('text_img.jpg')
```

# Bar Plot
---
- 데이터의 크기 비교
- bar()

```python
data = np.array([[5,25,50,20],[4,23,51,17],[6,22,52,19]])
x = np.arange(0,8,2)
```

```python
data
```

```python
# x = [0,2,4,6]
x = np.array([0,2,4,6])
```

```python
y = data[0]
y
```

```python
plt.bar(x,y, color= 'b', width = 0.5)
```

```python
plt.bar(x,data[0], color= 'b', width = 0.5)

# data[0]뒤에 data[1]그래프를 그려라
plt.bar(x,data[1], color= 'g', width = 0.5, bottom=np.sum(data[:0], axis=0))
# data[0] + data[1] 합친거 뒤에 data[2]그래프를 그려라
plt.bar(x,data[2], color= 'r', width = 0.5, bottom=np.sum(data[:1], axis=0))
```

```python
plt.bar(x, data[0], color='b', width=0.5)
# 가로축 방향으로 0.5만큼 이동해서 그래프 그리기
# x가 np.array일 경우 그냥 덧샘
# x가 list일 경우 더할 값을 리스트 * len(x)로 바꿔주기
plt.bar(x+0.5, data[1], color='g', width=0.5)
plt.bar(x+1, data[2], color='r', width=0.5)
```

# Box plot
---
- 데이터 분포를 표현
- boxplot()

```python
data = np.random.randn(100,5)
data[:5]
```

```python
plt.boxplot(data)

# box가 작을수록 데이터의 흩어짐이 적다
# 박스의 일정 범위 밖으로 나간 데이터 : outlier

# 최대값
# 상위 75% : 제 3사분위 수 (Q3)
# 주황색 선 :  제 2사분위 수 또는 중위수 (Q2)
# 하위 25% : 제 1사분위 수 (Q1)
# 최소값

plt.show()
```

# Pandas Visualization

```python
# plt.figure() - 도화지

# plt.plot() - 그래프

# plt.show() - 화면에 출력
```

```python
plt.figure(figsize=(10,5))

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
drinks = pd.read_csv(url)
drinks.head()
```

```python
drinks.columns = ['country', 'beer', 'spirit', 'wine', 'litres', 'continent']
drinks.head()
```

```python
df = drinks.copy()
```

```python
df.head()
```

```python
# df의 인덱스를 'country'로 변경 = df.set_index[0]
df = df.set_index('country')
df.head()
```

```python
# 기본 line chart
df.beer.plot()
```

```python
plt.figure(figsize = (15,3))
# 'beer'을 기준으로 정렬 후 'beer'의 row값을 출력
df.sort_values('beer')['beer'].plot()
```

```python
# bar 차트
plt.figure(figsize = (15,3))
df.beer.plot(kind='bar')
```

```python
plt.figure(figsize = (15,3))
# beer을 내림차순 정렬 후 상위 20개만 bar차트로 출력
df.beer.sort_values(ascending=False)[:20].plot(kind = 'bar')
```

```python
# 대륙별 음료 평균소비량
df.groupby('continent').mean().plot(kind='bar')
```

```python
# stacked : 겹쳐서
df.groupby('continent').mean().plot(kind='bar', stacked=True)
```

## histogram

```python
# 맥주를 많이 먹는 빈도수(0에 가까운 국가가 많다)
df['beer'].plot(kind='hist', bins=50)
```

## scatter

```python
# 전반적으로 와인소비량에 비해 맥주 소비량이 높다.
df.plot(x='beer',y='wine',kind='scatter')
```

```python
# .plotting.scatter_matrix : 모든 데이터(float로 구성된 데이터)간의 상관관계를 scatter로 보여준다
pd.plotting.scatter_matrix(df)
plt.show()
```

## boxplot

```python
df.plot(kind='box')
```

## matshow()

```python
# 칼럼간의 상관계수를 나타내준다
df.corr()
```

```python
# 색이 밝을 수록 상관관계가 높다
plt.matshow(df.corr())
```
