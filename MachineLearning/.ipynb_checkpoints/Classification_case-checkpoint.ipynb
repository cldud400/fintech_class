```python
df = pd.read_csv('./ml_source_files/santander_train.csv')
df.head()
```

```python
# Target => 불만족 : 1, 만족 : 0
```

```python
df.TARGET.value_counts()
```

```python
df.describe()
```

```python
# 이상치(outlier) 제거
df.var3.value_counts().sort_index()
df['var3'].replace(-999999, 2, inplace=True)
```

```python
# x, y data divide
```

```python
y = df['TARGET']
x = df.drop('TARGET', axis=1)
```

```python
x.drop('ID', axis=1, inplace=True)
```

```python
x.head(2)
```

```python
# train test split
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=111)
```

```python
# algorithm import
```

```python
from xgboost import XGBClassifier
```

```python
# eval_metric='auc' : auc곡선의 면적
# n_estimators : number of boosting round
xgb_clf = XGBClassifier(n_estimators=500, random_state=111)
# early_stopping_rounds : 더 이상 모델이 증진되지 않는 상태가 100 round이면 종료
xgb_clf.fit(x_train, y_train, early_stopping_rounds=100, eval_metric='auc',
           eval_set=[(x_train, y_train), (x_test, y_test)])

# xgboost : 모델이 순차적으로 만들어진다.
```

```python
# model evalutation
from sklearn.metrics import classification_report, roc_auc_score
```

```python
preds = xgb_clf.predict(x_test)
```

```python
print(classification_report(y_test, preds))
```

```python
# roc_auc_score : 확률값을 넣어줘야 한다.
roc_auc_score(y_test, xgb_clf.predict_proba(x_test)[:,1], average='macro')
```

```python
# GridSearchCV
```

```python
from sklearn.model_selection import GridSearchCV
# GridSearchCV 여러 매개변수를 돌려주는 클래스

xgb_cls = XGBClassifier(n_estimators=200)

# params = {'max_depth':[3, 4, 5],
#           'gamma':[0.5, 1, 1.5, 2,5],
#           'subsample':[0.6, 0.8, 1],
#           'min_child_weigh':[1, 5, 10], 
#           'colsample_bytree':[0.6, 0.8, 1.0]}
params = {'max_depth':[3,5],
         'colsample_bytree':[0.6,1.0]}

grid = GridSearchCV(xgb_cls, param_grid=params, verbose=False, cv=3)
grid.fit(x_train, y_train, early_stopping_rounds=30, eval_metric='auc',
        eval_set=[(x_train,y_train), (x_test,y_test)])
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
xgb_cls = XGBClassifier(n_estimators=200, colsample_tree=0.6, max_drpth=3)

xgb_cls.fit(x_train, y_train)
preds = xgb_cls.predict(x_test)
```

```python
print(classification_report(y_test, preds))
```

```python
roc_auc_score(y_test, xgb_cls.predict_proba(x_test)[:,1], average='macro')
```
