```python
df = pd.read_csv('eighthr.data', header=None)
df.head()
```

```python
# neural network로 예측하는 모델, LSTM으로 예측하는 모델을 각각 생성하고 평가하시오..

# target : 73번 칼럼
# object => numeric
# scaling
```

```python
df = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')
df.info()
```

```python
df.isnull().sum()
```

```python
df
```

```python
df1 = df.dropna()
df1.info()
```

```python
scaler = MinMaxScaler()
df1_sc = scaler.fit_transform(df1)
df1_sc
```

```python
y = df1_sc[:, -1]
x = df1_sc[:,:-1]
```

```python
y
```

```python
x.shape
```

```python
x = x.reshape(1847, 72, 1)
```

```python
model = Sequential()
model.add(LSTM(128, input_shape=(72,1)))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=20, batch_size=10, validation_split=0.2)
```
