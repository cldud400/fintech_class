```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./datafiles/housing.csv', header=None, delim_whitespace=True)
df.head()
```

```python
df.info()
```

```python
y = df.values[:,13]
x = df.values[:,:13]
```

```python
# 연속 값을 예측하는 것 continue data => regression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=111)
```

```python
x_train.shape
```

```python
# 모델 생성
```

```python
model = Sequential()
# model에 layer을 add해준다.
# input layer에 차원을 맞춰준다
model.add(Dense(30, input_dim=13, activation='relu'))
# hiden layer
model.add(Dense(20, activation='relu'))
model.add(Dense(10, activation='relu'))
# output layer
# linear regression에서는 output layer에 activation function을 사용하지 않는다.
model.add(Dense(1))

# binary cls activation : sigmoid()
# multiclassification : softmax()
# linear regression : None
```

```python
# model compile
```

```python
# 회귀모델에서의 평가지표는 MSE, r2_score
# MSE가 loss를 측정하는데도 쓰이고, 평가를 하는데도 쓰인다.
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
# epochs : 학습을 시키는 횟수
# history로 학습데이터를 저장할 수 있다.
# validation_split : 주어진 데이터셋을 비율로 나눠서 트레이닝 데이터셋과 테스트 데이터셋으로 지정 (검증 용도)
hist = model.fit(x_train, y_train, validation_split=0.2, epochs=200, batch_size=20)
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```

```python
model.evaluate(x_test, y_test)
```

```python
# scaling
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model = Sequential()
model.add(Dense(30, input_dim=13, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
hist = model.fit(x_train_scaled, y_train, validation_split=0.2, epochs=200, batch_size=20)
```

```python
model.evaluate(x_test_scaled, y_test)
```

```python
# prediction
```

```python
preds = model.predict(x_test_scaled)
```

```python
# evaluation
```

```python
from sklearn.metrics import r2_score

r2_score(preds, y_test)
```

```python
for i in range(10):
    pred_price = preds[i]
    real_price = y_test[i]
    print(f'{round(real_price)} - {round(pred_price[0])}')
```
